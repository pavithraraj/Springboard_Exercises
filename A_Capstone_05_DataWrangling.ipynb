```python
# AMES- IOWA Housing data - Basic data wrangling steps 
import pandas as pd
df1 = pd.read_csv('train.csv', header=0)
#print(df1)
df1.head(10)
```

```python
# GRoup the variables by LotShape and then check for mean, max 
df1_yr_sold = df1.groupby('LotShape')
countbylotshape = df1_yr_sold['LotShape'].count()
df1_yr_sold['SalePrice'].describe()
```

```python
df1.shape
```

```python
# Initial data check to see if there # of houses that fall is a particular lotShape and MSZoning
# RL in  Zoning type has MAX count for any LotShape
df1_two_columns = df1.groupby(['LotShape','MSZoning'])
df1_two_columns['SalePrice'].count()
df1_two_columns['SalePrice'].describe()
```

```python
# In each groups for LotShape and MSZoning find the max value that correlated to SalePrice
# We see that RL Zoning has the Maximum SalePrice
df1_two_columns['SalePrice'].max()
```

```python
# Check if the SalePrice has corelation on the Month Sold
# From the scatter plot it seems that the houses are below 400k range
# We see that Month Sold not very significant co-relation so exploring what other variables influcence SalesPrice
import matplotlib.pyplot as plt
df1.plot(kind='scatter', x='SalePrice', y='MoSold', rot=70)
plt.show()
```
